```python
import cv2
#opencv library
```

```python
configPath='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
weightsPath='frozen_inference_graph.pb'
```

setting the configuration and scale

```python
net = cv2.dnn_DetectionModel(weightsPath,configPath)
net.setInputSize(320,320)
net.setInputScale(1.0/127.5)
net.setInputMean((127.5,127.5,127.5))
net.setInputSwapRB(True)
```

setting the coco namess classes for evaluation of objects

```python
classNames=[]
classFile='coco.names'
with open(classFile,'rt') as f:
    classNames = f.read().rstrip('\n').split('\n')
classNames    
```

trying the code on the image first

```python
img = cv2.imread('face.png')
#cv2.imshow("Output",img)
#cv2.waitKey(0)
```

```python
classIds,confs,bbox =  net.detect(img, confThreshold=0.5)
print(classIds,bbox)
```

```python
for classId,confidence,box in zip(classIds.flatten(),confs.flatten(),bbox):
    cv2.rectangle(img,box,color=(0,255,0),thickness=2)
    cv2.putText(img,classNames[classId-1],(box[0]+10,box[1]+30),cv2.FONT_HERSHEY_PLAIN,1,(0,255,0),2)
```

```python
#cv2.imshow("Output",img)
#cv2.waitKey(0)
```

setting up the camera for the object detection

```python
cap=cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)
```

entering threshold 0.6, so that the detection with accuracy reached is 60% or more

```python
thresh=0.6
```

```python
cap.read()
```

```python
while True:
    success,img= cap.read()
    classIds,confs,bbox =  net.detect(img, confThreshold=thresh)
    print(classIds,bbox)
    
    if len(classIds)!=0:
        for classId,confidence,box in zip(classIds.flatten(),confs.flatten(),bbox):
            cv2.rectangle(img,box,color=(0,255,0),thickness=2)
            cv2.putText(img,classNames[classId-1],(box[0]+10,box[1]+30),cv2.FONT_HERSHEY_PLAIN,1,(47, 0, 255),2)
            cv2.putText(img,str(confidence),(box[0]+10,box[1]+50),cv2.FONT_HERSHEY_PLAIN,1,(47, 0, 255),2)
    cv2.imshow("Output",img)
    cv2.waitKey(1)
    if cv2.waitKey(1) == ord("q"): # check if the user pressed "q" 
        break
cap.release()
cv2.destroyAllWindows()
```
